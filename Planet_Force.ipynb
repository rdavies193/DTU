```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [20, 12]

G=6.67e-11
msun=2.0e30

fig, ax = plt.subplots()

planet=["Mercury","Venus","Earth","Mars","Jupiter","Saturn","Uranus","Neptune"]
mass=[0.330,4.87,5.97,0.642,1898.,568.,86.8,102.]
distance=[57.9,108.2,149.6,227.9,778.6,1433.5,2872.5,4495.1]

mass_kg = [i * 1e24 for i in mass] ## kg 
dist_1e9= [i * 1e9 for i in distance] ## metres
force = [G*msun*a/b**2 for a, b in zip(mass_kg, dist_1e9)] ## N

color = ['lightcoral', 'darkorange', 'olive', 'teal', 'violet', 
         'skyblue','red','cornflowerblue']

ax.grid(True, linestyle='-.')
ax.tick_params(width=3,labelsize=16)
plt.scatter(distance,force,s=200,c=color)
fs=18
plt.xlabel('Distance (million km)',fontsize=fs)
plt.ylabel('Force (N)',fontsize=fs)
plt.show()
fig.savefig('Planet_Force.jpg', bbox_inches = "tight",dpi=500)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = [20, 12]

G=6.67e-11
msun=2.0e30

fig, ax = plt.subplots()
scatter_x = np.array([0.330,4.87,5.97,0.642,1898.,568.,86.8,102.])  ##mass
scatter_y = np.array([57.9,108.2,149.6,227.9,778.6,1433.5,2872.5,4495.1]) ##distance

group = np.array(["Mercury","Venus","Earth","Mars","Jupiter","Saturn","Uranus","Neptune"])
mass_kg = [i * 1e24 for i in mass] ## kg 
dist_1e9= [i * 1e9 for i in distance] ## metres
force = [G*msun*a/b**2 for a, b in zip(mass_kg, dist_1e9)] ## N
force_ar = np.array(force)

for g in np.unique(group):
    i = np.where(group == g)
    ax.scatter(scatter_y[i], force_ar[i], label=g,s=200)

fs=18
ax.legend(fontsize=fs)
ax.grid(True, linestyle='-.')
ax.tick_params(width=3,labelsize=16)
plt.xlabel('Distance (million km)',fontsize=fs)
plt.ylabel('Force (N)',fontsize=fs)

plt.show()
```
